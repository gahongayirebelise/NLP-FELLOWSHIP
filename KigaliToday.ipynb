```python
from bs4 import BeautifulSoup
import bs4 
import requests
import pandas as pd
```

```python
content_info = requests.get("https://www.kigalitoday.com/amakuru/").content
Soup = BeautifulSoup(content_info, "html.parser")
web_info = Soup.find_all('div', class_= "headline-container")
```

```python
df = pd.DataFrame()
df['link'] = None
df['title'] = None
df['Description'] = None
df['Article_date'] = None
```

```python
df
```

```python
def my_function(web_info):
  link = []
  title = []
  Description = []
  Article_date = []
  for index, item in enumerate(web_info):
    link.append(item.find('a')['href'])
    title.append(item.find('h3').find('a'))
    Description.append(item.find('div').find('p').getText())
    Article_date.append(item.find('span', class_= "article_date"))
  return link, title, Description, Article_date
```

```python
df['link'], df['title'], df['Description'], df['Article_date'] = my_function(web_info)
```
