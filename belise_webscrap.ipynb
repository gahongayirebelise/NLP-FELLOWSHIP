```python
from bs4 import BeautifulSoup as bs
```

```python
import bs4
import requests as r
```

```python
html_text = r.get("https://web.archive.org/web/20220201000813/https://www.igihe.com/")
Soup = bs(html_text.text, "html.parser")
print(Soup.prettify())
```

```python
text_file= Soup.find_all('span', class_='homenews-title')
text_file[0].find('a')['href']
```

```python
# finding all links
links = []
for span in text_file:
  links.append(span.find('a'))
  print(links)
```

```python
# finding link titles
link_titles = []
for link in links:
  link_titles.append(link["href"])
link_titles
```

```python
#urls and titles
url_and_title = []
for title in link_titles:
  url_and_title.append(+title)
url_and_title
```
